** task-2 create a datframe**

```python
lables=visa_df['continent'].unique()

applicants=[]

for i in lables:
    con=visa_df['continent']==i
    count=len(visa_df[con])
    applicants.append(count)

applicants

```

```python
data=pd.DataFrame(zip(lables,applicants),columns=['continent','count'])
data.to_csv('Continent.csv')
```

```python
lables1=visa_df['case_status'].unique()


applicants=[]

for i in lables1:
    con=visa_df['case_status']==i
    count1=len(visa_df[con])
    applicants.append(count1)

applicants

case_status=pd.DataFrame(zip(lables1,applicants),columns=['status','count'])
case_status
case_status.to_csv('case_status.csv')
```

**bar charts**

- We can plot the bar charts for catagorical columns
- Bar charts requires class as x_axis and class frequecy as y-axis
- We have a method under **matplotlib: plt.bar**
   - need keys and values count
   - it also need data frame name

```python
continent_vs=visa_df['continent'].value_counts()
keys=continent_vs.keys()
values=continent_vs.values

plt.bar(keys,values)
```

```python
continent_vs=visa_df['continent'].value_counts()
keys=continent_vs.keys()
values=continent_vs.values

colurs=['red','green','yellow','pink','orange','grey']
plt.figure(figsize=(10,6))
plt.bar(keys,values,color=colurs)
```

```python
continent_vs=visa_df['continent'].value_counts()
keys=continent_vs.keys()
values=continent_vs.values

colurs=['red','green','yellow','pink','orange','grey']
plt.figure(figsize=(10,6))
plt.bar(keys,values,color=colurs)
plt.xlabel('Continents')
plt.ylabel('num of applicants')
plt.title('continent bar charts')
plt.show()
# plt.savefig('continent_bar_charts.jpg')
```
